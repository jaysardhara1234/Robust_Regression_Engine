### Q1 . What is Regularization in Machine Learning ? Why is it needed ?

### Regularization is a set of technique used to reduce overfitting in machine learning models by adding a penalty term to the cost function. This penalty discourages very large coefficient values and prevents the model from becoming too complex.This helps the model perform better on unseen data.

### Q2 . Difference between Lasso Regression (L1) and Ridge Regression (L2).

### Lasso uses Absolute values of coefficients for penalty while, Ridge uses Squared values of coefficients for penalty.

### Lasso can make some coefficients exactly Zero while, Ridge Reduces coefficients but does not make them Zero.
### Lasso performs feature selection while, Ridge keeps all features in the model.

### Q3 . What is Cross-Validation and why is it important ?

### Cross-Validation is a model evaluation technique in which the dataset is divided into multiple subsets, and the model is trained and tested on different combinations of these subsets. It provides a more reliable estimate of model performance and helps detect overfitting and underfitting.

### Q4 . Explain the following Cross-Validation Techniques :

### A . K-Fold Cross Validation : The dataset is divided into K equal parts (folds). The model is trained on K−1 folds and tested on the remaining fold. This process is repeated K times, and the average score is calculated.

### B . Stratified K-Fold Cross Validation : Stratified K-Fold ensures that each fold contains approximately the same proportion of target classes as the original dataset. It is mainly used for classification problems with imbalanced classes.

### C . Leave-One-Out Cross Validation (LOOCV) : In LOOCV, one observation is used for testing and all remaining observations are used for training. This process is repeated for every observation in the dataset but is computationally expensive.

### D . Time Series Split : Time Series Split is used for time-based data. It keeps the chronological order of the data by training on past data and testing on future data.

### Q5 . Why are tree-based models less sensitive to feature scaling ?

### Tree-based models split data based on feature values and decision rules rather than distance calculations. Since scaling does not change the order of feature values, Therefore, feature scaling is generally not required for tree-based models.
